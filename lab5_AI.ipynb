```python
import numpy as np
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import LabelEncoder

# Step 1: Define the data as numpy arrays (without pandas)
data = np.array([
    ['male', 6.00, 180, 12],
    ['male', 5.92, 190, 11],
    ['male', 5.58, 170, 12],
    ['male', 5.92, 165, 10],
    ['female', 5.00, 100, 6],
    ['female', 5.50, 150, 8],
    ['female', 5.42, 130, 7],
    ['female', 5.75, 150, 9]
])

# Step 2: Encode the 'Gender' column (Male -> 1, Female -> 0)
label_encoder = LabelEncoder()
gender_encoded = label_encoder.fit_transform(data[:, 0])  # Encoding 'male' to 1 and 'female' to 0

# Step 3: Prepare the features (X) and target variable (y)
X = data[:, 1:].astype(float)  # Features: Height, Weight, Foot_Size (convert to float)
y = gender_encoded  # Target: Encoded Gender (0 or 1)

# Step 4: Train the Decision Tree classifier
clf = DecisionTreeClassifier(random_state=42)
clf.fit(X, y)

# Step 5: Define a new entry (example input)
new_entry = np.array([[5.5, 160, 10]])  # New input entry: Height=5.5, Weight=160, Foot_Size=10

# Step 6: Make the prediction using the trained classifier
predicted_gender = clf.predict(new_entry)[0]

# Step 7: Convert the predicted gender (0 or 1) back to the gender label (male or female)
predicted_gender_label = label_encoder.inverse_transform([predicted_gender])[0]

# Step 8: Output the prediction result for the new entry
print(f"Predicted gender for the new entry (Height=5.5, Weight=160, Foot_Size=10): {predicted_gender_label}")

# Optional: Add the new entry back to the dataset (with predicted gender)
new_entry_with_gender = np.hstack([new_entry, np.array([[predicted_gender]])])  # Add predicted gender to the new entry
updated_data = np.vstack([data, new_entry_with_gender])  # Append the new entry to the dataset

# Step 9: Retrain the classifier with the updated dataset
X_updated = updated_data[:, 1:].astype(float)  # Updated features (Height, Weight, Foot_Size)
y_updated = label_encoder.transform(updated_data[:, 0])  # Updated target (Gender)
clf.fit(X_updated, y_updated)

# Step 10: Display the updated dataset
print("\nUpdated dataset after adding the new entry:")
print(updated_data)

```

```python
import numpy as np
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import LabelEncoder

# Step 1: Define the data as numpy arrays (without pandas)
data = {
    'Gender': ['male', 'male', 'male', 'male', 'female', 'female', 'female', 'female'],
    'Height': [6.00, 5.92, 5.58, 5.92, 5.00, 5.50, 5.42, 5.75],
    'Weight': [180, 190, 170, 165, 100, 150, 130, 150],
    'Foot_Size': [12, 11, 12, 10, 6, 8, 7, 9]
}

# Step 2: Convert data into numpy arrays for processing
gender = np.array(data['Gender'])
height = np.array(data['Height'])
weight = np.array(data['Weight'])
foot_size = np.array(data['Foot_Size'])

# Step 3: Encode the 'Gender' column (Male -> 1, Female -> 0)
label_encoder = LabelEncoder()
gender_encoded = label_encoder.fit_transform(gender)  # Encoding 'male' to 1 and 'female' to 0

# Step 4: Prepare the features (X) and target variable (y)
X = np.column_stack((height, weight, foot_size))  # Features: Height, Weight, Foot_Size
y = gender_encoded  # Target: Encoded Gender (0 or 1)

# Step 5: Train the Decision Tree classifier
clf = DecisionTreeClassifier(random_state=42)
clf.fit(X, y)

# Step 6: Define a new entry (example input)
new_entry = np.array([[5.5, 160, 10]])  # New input entry: Height=5.5, Weight=160, Foot_Size=10

# Step 7: Make the prediction using the trained classifier
predicted_gender = clf.predict(new_entry)[0]

# Step 8: Convert the predicted gender (0 or 1) back to the gender label (male or female)
predicted_gender_label = label_encoder.inverse_transform([predicted_gender])[0]

# Step 9: Output the prediction result for the new entry
print(f"Predicted gender for the new entry (Height=5.5, Weight=160, Foot_Size=10): {predicted_gender_label}")

# Optional: Add the new entry back to the dataset (with predicted gender)
new_entry_with_gender = np.hstack([new_entry, np.array([[predicted_gender]])])  # Add predicted gender to the new entry
updated_data = np.vstack([np.column_stack((height, weight, foot_size, gender_encoded)), new_entry_with_gender])  # Append the new entry to the dataset

# Step 10: Retrain the classifier with the updated dataset
X_updated = updated_data[:, :-1]  # Updated features (Height, Weight, Foot_Size)
y_updated = updated_data[:, -1]  # Updated target (Gender)
clf.fit(X_updated, y_updated)

# Step 11: Display the updated dataset (just for visualization)
print("\nUpdated dataset after adding the new entry:")
print(updated_data)
```
